```python
import torch
import numpy as np

print(torch.__version__)
print(numpy.__version__)
```

```python
tensor = torch.rand(4,3) # Create random 4 by 3 torch Tensor
tensor
```

```python
# We can verify something is torch tensor by using type method
type(tensor)
```

```python
# Torch Tensors are NumPy Arrays with the ability to be executed on GPUs
numpy_from_tensor = tensor.numpy()
numpy_from_tensor
```

```python
type(numpy_from_tensor) # Now we can see numpy type
```

```python
# Can also use to check in code the torch.is_tensor() method ex:
torch.is_tensor(tensor)
```

```python
torch.is_tensor(numpy_from_tensor)
```

```python
# The NumPy array and the PyTorch tensor share the same underlying memory
numpy_from_tensor[0, 0] = 100.0
numpy_from_tensor
```

```python
tensor # See the first element was also changed her
```

```python
# Now let's do the reverse and go from numpy arr to tensor
numpy_arr = np.array([[1.0, 2.0, 3.0],
                      [10.0, 20.0, 30.0],
                      [100.0, 200.0, 300.0]])
numpy_arr
```

```python
tensor_from_numpy = torch.from_numpy(numpy_arr)
tensor_from_numpy
```

```python
type(tensor_from_numpy)
```

```python
torch.is_tensor(tensor_from_numpy)
```

```python
# Changes to tensor are still reflected in numpy array since still same memory
```

```python
# Torch allows you to initiate from numpy in others way too
np_array_one = np.array([4, 8])
np_array_one
```

```python
tensor_from_array_one = torch.as_tensor(np_array_one)
# Performs a copy if the data is not already a tensor
tensor_from_array_one
```

```python
np_array_one[1] = 5
tensor_from_array_one
# We can see it is still updated when original np array was changed
```

```python
# We can also create a copy to create a new tensor
np_array_two = np.array([2, 2])
tensor_from_array_two = torch.tensor(np_array_two)
tensor_from_array_two
```

```python
np_array_two[1] = 4
tensor_from_array_two # see it did not get updated!
```
